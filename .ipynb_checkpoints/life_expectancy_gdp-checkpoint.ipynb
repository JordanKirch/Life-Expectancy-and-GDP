```python
import pandas as pd
import numpy as np
import seaborn as sns
import statsmodels
import matplotlib.pyplot as plt
import math
```

```python
df = pd.read_csv('all_data.csv')
```

```python
print(df.head())
```

```python
# Which Country has the highest Life Expectancy
sns.barplot(data = df, x = 'Country', y = 'Life expectancy at birth (years)')
plt.show()
plt.clf()
print('Mean life expectancy: ' + str(df['Life expectancy at birth (years)'].mean()))
```

```python
# Has life expectancy increased over time in the six nations?
sns.lineplot(data = df, x = 'Year', y = 'Life expectancy at birth (years)', hue = 'Country')
plt.show()
plt.clf()
# Chile, China, Germany, Mexico, and USA all have stable linear growth of life expectancy.
# Zimbabwe on the other hand, has seen dramatic increase in life expectancy since 2004
```

```python
# Has GDP increased over time in the six nations?
sns.lineplot(data = df, x = 'Year', y = 'GDP', hue = 'Country')
plt.show()
plt.clf()
# USA and China Have seen the greatest increases in GDP over the years
```

```python
# Is there a correlation between GDP and life expectancy of a country?
sns.scatterplot(data = df, x = 'Life expectancy at birth (years)', y = 'GDP', hue = 'Country')
plt.show()
plt.clf
# Zimbabwe sample
zimbabwe_df = df[df['Country'] == 'Zimbabwe']
sns.scatterplot(data = zimbabwe_df, x = 'Life expectancy at birth (years)', y = 'GDP')
plt.legend(['Zimbabwe'])
plt.show()
plt.clf

# There is a positive correlation between GDP and Life expctancy
```
